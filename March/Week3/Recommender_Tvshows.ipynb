# Recommendation

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
cv
```

```python
vector=cv.fit_transform(tv_shows_1['description.1'].values.astype('U')).toarray()
vector.shape
```

```python
similarity=cosine_similarity(vector)
similarity
```

```python
tv_shows_1[tv_shows_1['season_name']=="Batman"].index[0]
```

```python
# Ensure tv_shows_1 is a DataFrame
assert isinstance(tv_shows_1, pd.DataFrame), "tv_shows_1 is not a DataFrame"

distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector: vector[1])
for i in distance[0:5]:
    # Ensure i[0] is a valid index for tv_shows_1
    if i[0] < len(tv_shows_1):
        print(tv_shows_1.iloc[i[0]]['season_name'])  # Access the 'season_name' attribute using dictionary-like indexing
    else:
        print("Index out of bounds:", i[0])
```

```python
def recommend(series):
    # Check if any rows match the condition
    if not tv_shows_1[tv_shows_1['season_name'] == series].empty:
        index = tv_shows_1[tv_shows_1['season_name'] == series].index[0]
        distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])
        for i in distance[0:5]:
            print(tv_shows_1.iloc[i[0]]['season_name'])  # Access the 'season_name' attribute using dictionary-like indexing
    else:
        print("No matching series found for:", series)
```

```python
# Call the function with the appropriate argument
recommend('Instant Mom: Volume 1')
```
