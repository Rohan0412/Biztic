# Proper 1

```python
import sqlite3
import csv
import requests
from bs4 import BeautifulSoup

def create_database_table():
    connection = sqlite3.connect('scraped_data.db')
    cursor = connection.cursor()

    cursor.execute('''
        CREATE TABLE IF NOT EXISTS products (
            id INTEGER PRIMARY KEY,
            url TEXT
        )
    ''')

    connection.commit()
    connection.close()

def insert_into_database(product_url):
    connection = sqlite3.connect('scraped_data.db')
    cursor = connection.cursor()

    cursor.execute('''
        INSERT INTO products (url)
        VALUES (?)
    ''', (product_url,))

    connection.commit()
    connection.close()

def export_to_csv(csv_file):
    connection = sqlite3.connect('scraped_data.db')
    cursor = connection.cursor()

    cursor.execute('SELECT * FROM products')
    rows = cursor.fetchall()

    with open(csv_file, 'w', newline='') as file:
        csv_writer = csv.writer(file)
        csv_writer.writerow(['id', 'url'])
        csv_writer.writerows(rows)

    connection.close()

def scrape_and_store_data(url):
    response = requests.get(url)
    if response.status_code == 200:
        html_code = response.text
        soup = BeautifulSoup(html_code, 'html.parser')

        if soup:
            create_database_table()

            product_containers = soup.find_all('ul', class_='megamenu')
            
            for product_container in product_containers:
                product_urls = [li.find('a')['href'] for li in product_container.find_all('li')]
                for product_url in product_urls:
                    insert_into_database(product_url)
                    
                    
            export_to_csv('exported_data.csv')
        else:
            print("Error: 'soup' object is None.")
    else:
        print(f"Error: Failed to fetch the page. Status code: {response.status_code}")

# URL of the website to scrape
url_to_scrape = 'https://mdcomputers.in/'

# Call the function to scrape and store data
scrape_and_store_data(url_to_scrape)
```
