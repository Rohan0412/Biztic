```python
import requests
from bs4 import BeautifulSoup

def scrape_product_details(product):
    title = product.find('h4').find('a').text.strip()
    price_old = product.find('span', class_='price-old').text.strip()
    price_new = product.find('span', class_='price-new').text.strip()
    discount = product.find('span', class_='label-product label-sale').text.strip()

    print(f"Title: {title}")
    print(f"Old Price: {price_old}")
    print(f"New Price: {price_new}")
    print(f"Discount: {discount}")
    print('\n' + '-'*50 + '\n')

def scrape_product_urls(url, max_pages=None):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    try:
        for page in range(1, max_pages + 1) if max_pages else range(1, float('inf')):
            current_url = f"{url}/all-products?page={page}"
            response = requests.get(current_url, headers=headers)

            if response.status_code == 200:
                soup = BeautifulSoup(response.text, 'html.parser')

                # Find all product containers on the page
                product_containers = soup.find_all('div', class_='product-item-container')  # Replace with the actual class for each product

                for product_container in product_containers:
                    # Find all products within the container
                    products = product_container.find_all('div', class_='product-item-container')  # Replace with the actual class for each individual product

                    for product in products:
                        product_url = product.find('h4').find('a')['href']
                        print(f"\nURL: {product_url} \n{'-'*30}")

                        # Call the function to scrape details from the individual product page
                        scrape_product_details(product)

                next_page = soup.find('li', class_='next')
                if not next_page:
                    break  # Exit the loop if there is no next page

            else:
                print(f"Failed to fetch the page {current_url}. Status code: {response.status_code}")
                break

    except Exception as e:
        print(f"An error occurred while processing page {current_url}: {e}")

# Example usage
url_to_scrape = 'https://mdcomputers.in'
max_pages_to_scrape = 3  # Adjust the number of pages to scrape or set to None for all pages
scrape_product_urls(url_to_scrape, max_pages=max_pages_to_scrape)
```
