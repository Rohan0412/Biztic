```python
def get_html_code(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Raises HTTPError for bad requests
        return response.text
    except requests.exceptions.RequestException as e:
        print(f"Failed to fetch the page. Error: {e}")
        return None
```

```python
def get_all_links(html_code,url):
    soup = BeautifulSoup(html_code, 'html.parser')
    product_containers = soup.find_all('ul', class_='megamenu')
    main_links = []

    for product_container in product_containers:
        product_urls = [urljoin(url, li.find('a')['href']) for li in product_container.find_all('li')]
        main_links.extend(product_urls)

    if not product_urls:
        print("No product container found.")

    return product_urls
```

```python
def scrape_product_details(url):
    html_code = get_html_code(url)

    if html_code:
        soup = BeautifulSoup(html_code, 'html.parser')

        
        title_elem = soup.find('h4').find('a')
        title = title_elem.text.strip() if title_elem else 'Title not found'

        
        product_url_elem = soup.find('h4').find('a')
        product_url = product_url_elem['href'].strip() if product_url_elem else 'URL not found'

        
        price_old_elem = soup.find('span', class_='price-old')
        price_old = price_old_elem.text.strip() if price_old_elem else 'Old Price not found'

        
        price_new_elem = soup.find('span', class_='price-new')
        price_new = price_new_elem.text.strip() if price_new_elem else 'New Price not found'

        
        discount_elem = soup.find('span', class_='label-product label-sale')
        discount = discount_elem.text.strip() if discount_elem else 'Discount not found'

        print(f"Title: {title}")
        print(f"URL: {product_url}")
        print(f"Old Price: {price_old}")
        print(f"New Price: {price_new}")
        print(f"Discount: {discount}")
    else:
        print("Error: HTML code is not available.")
```

```python
def scrape_and_store_data_with_pagination_and_link(base_url):
    page_num = 1
    while True:
        page_url = f"{base_url}?page={page_num}"  # Assuming pagination is based on query parameter 'page'
        html_code = get_html_code(page_url)

        if html_code:
            product_urls = get_all_links(html_code, base_url)
            if not product_urls:
                print(f"No product links found on page {page_num}. Exiting pagination.")
                break

            for product_url in product_urls:
                scrape_product_details(product_url)

            page_num += 1
        else:
            print(f"Error: HTML code for page {page_num} is not available.")
            break

    
    new_link = get_next_link(html_code)  # Implement a function to extract the next link
    if new_link:
        print(f"Moving to the next link: {new_link}")
        scrape_and_store_data_with_pagination_and_link(new_link)
    else:
        print("No more links to scrape.")
```

```python
url_to_scrape = 'https://mdcomputers.in/'  
scrape_and_store_data_with_pagination_and_link(url_to_scrape)
```

```python
def get_all_links(html_code, url):
    soup = BeautifulSoup(html_code, 'html.parser')
    product_containers = soup.find_all('ul', class_='megamenu')
    main_links = []

    for product_container in product_containers:
        product_urls = [urljoin(url, li.find('a')['href']) for li in product_container.find_all('li')]
        main_links.extend(product_urls)

    if not main_links:
        print("No product links found.")

    return main_links
```

```python
def store_in_database(data_list):
    conn = mysql.connector.connect(
        host='localhost',
        user='root',
        password='1234',
        database='md'
    )
    cursor = conn.cursor()

    cursor.execute('''
        CREATE TABLE IF NOT EXISTS products (
            id INT AUTO_INCREMENT PRIMARY KEY,
            title VARCHAR(250),
            url VARCHAR(250),
            price_old VARCHAR(250),
            price_new VARCHAR(250),
            discount VARCHAR(250)
        )
    ''')

    cursor.executemany('''
        INSERT INTO products (title, url, price_old, price_new, discount)
        VALUES (%s, %s, %s, %s, %s)
    ''', data_list)

    conn.commit()
    conn.close()
```

```python
def scrape_product_details(url):
    html_code = get_html_code(url)

    if html_code:
        soup = BeautifulSoup(html_code, 'html.parser')

        title_elem = soup.find('h4').find('a')
        title = title_elem.text.strip() if title_elem else 'Title not found'

        product_url_elem = soup.find('h4').find('a')
        product_url = product_url_elem['href'].strip() if product_url_elem else 'URL not found'

        price_old_elem = soup.find('span', class_='price-old')
        price_old = price_old_elem.text.strip() if price_old_elem else 'Old Price not found'

        price_new_elem = soup.find('span', class_='price-new')
        price_new = price_new_elem.text.strip() if price_new_elem else 'New Price not found'

        discount_elem = soup.find('span', class_='label-product label-sale')
        discount = discount_elem.text.strip() if discount_elem else 'Discount not found'
        
        product_data = (title, product_url, price_old, price_new, discount)

        store_in_database([product_data])
    else:
        print("Error: HTML code is not available.")
        return None
```
